```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.concat(
    map(
        pd.read_csv,
        [
            'data/play_by_play_2020.csv', 'data/play_by_play_2021.csv',
            'data/play_by_play_2022.csv', 'data/play_by_play_2023.csv', 'data/play_by_play_2024.csv'
        ]
    ),ignore_index=True
)
```

```python
df
```

```python
df.sample(7)
```

```python
rows, columns = df.shape
print(df.shape)
print(f"Liczba wierszy: {rows}\nLiczba column: {columns}")
```

Połączenie danych z sezonów rozgrywek NFL z pięciu poprzednich lat w jedną bazę danych.

```python
nfl_features = [
    'season', 'passer_id', 'passer',
    'pass', 'complete_pass', 'interception',
    'sack', 'yards_gained', 'touchdown'
]

groupby_features = ['season', 'passer_id', 'passer']

nfl_df = (df
          .loc[:, nfl_features]
          .groupby(groupby_features,
                   as_index=False)
          .sum())

nfl_df.head(7)
```

Uzwględnienie ważnych cech takich jak sezon id podającego, nazwa podającego, podania, przechwyty podan, oraz ataki na rozgrywającego aby nie dopuścić do dobrego podania przyczyniającego się do przyłożenia. Pogrupowanie ich według najważniejszych danych.

```python
nfl_df.corr(numeric_only=True)
```

Jak można zauważyć ilość podań, ukończone podania, przechwyty, ataki na rozgrywającego oraz zdobyte yardy mają wysoką korelacje z przyłożeniami.

```python
for data in ['pass', 'complete_pass',
            'interception', 'sack', 'yards_gained']:
    sns.regplot(data=nfl_df, x='touchdown', y=data, line_kws={'color':'red'})
    plt.title(f"touchdowns and {data}")
    plt.show()
```

Wizualizacja danych korelujących ze sobą, sprawdzanie za pomocą linii regresji przechodzącej przez punkty danych.

# Linear model use sklearn library
1. Let's find down how work simply Linear Reggresion


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

```python
nfl_df_copy = nfl_df.copy() 
```

```python
ml_features = [
    'pass', 'complete_pass', 'interception', 
    'sack', 'yards_gained'
]
target = 'touchdown'

model_data = (nfl_df_copy.dropna(subset=ml_features+[target]))

model_data.isnull().any()
```

Usunięcie wartości nullowych oraz sprawdzenie czy wszystko poszło zgodnie z myślą.

```python
train_data = (model_data.loc[model_data["season"] < 2024])
test_data = (model_data.loc[model_data["season"] == 2024])
```

```python
train_data
```

```python
test_data
```

Podział danych na trenowane i testowe, ukrycie ostatniego sezonu przed modelem aby później móc porównać czy dobrze przewiduje. Rzut okiem aby sprawdzić czy wszystko się zgadza.

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(train_data.loc[:, ml_features], train_data[target])
```

```python
preds = model_lr.predict(test_data.loc[:, ml_features])
preds
```

Uzyskanie przewidywań

```python
preds = pd.Series(preds, index=test_data.index)
preds
```

Przekonwertowanie tych przewidywań na bardziej zrozumiałe liczby dla ludzi

```python
test_data['preds'] = preds
test_data.sample(7)
```

Dodanie kolumny przewidywań do testowego zbioru i porownanie wizualnie przewidywania modelu.

```python
rmse = np.sqrt(mean_squared_error(test_data['touchdown'], test_data['preds']))
print(f"Średni błąd kwadratowy wynosi: {rmse} \nWychodzi na to, że model średnio myli się około 2 przyłożenia")
```

```python
sns.regplot(data=test_data, x="touchdown", y="preds", line_kws={'color': 'yellow'})
plt.title("touchdown and prediction")
plt.show()
```
